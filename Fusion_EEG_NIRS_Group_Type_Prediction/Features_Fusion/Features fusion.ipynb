```python
 # =========================
#  EEG + NIRS + Fusion CV
#  (Classiques + GRU + BiLSTM)
# =========================
import os
import numpy as np
import pandas as pd
import warnings
import logging

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, roc_auc_score
)

# Classiques
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier, log_evaluation, reset_parameter

# Signal/Features
from scipy import signal
from scipy.stats import entropy
from scipy.signal import welch

# Deep
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Bidirectional, LSTM, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Visu (optionnel)
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.exceptions import UndefinedMetricWarning
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)
logging.getLogger('lightgbm').setLevel(logging.WARNING)

# =========================
#      Chemins datasets
# =========================
eeg_path = "C:/Users/HP/Desktop/Dataset-EEG"
nirs_path = "C:/Users/HP/Desktop/Dataset-NIRS"

# =========================
#   Chargement / Appariement
# =========================
def load_data(path, file_prefix):
    files = sorted([f for f in os.listdir(path) if f.startswith(file_prefix)])
    data, labels = [], []
    for file in files:
        df = pd.read_csv(os.path.join(path, file), low_memory=False)
        data.append(df)
        labels.append(1 if file.startswith('AO') else 0)
    return files, data, labels

# EEG
eeg_files, eeg_data, eeg_labels = load_data(eeg_path, ('AO', 'SO'))
# NIRS
nirs_files, nirs_data, nirs_labels = load_data(nirs_path, ('AO', 'SO'))

# Apparier EEG <-> NIRS
matched_data = []
for i, eeg_file in enumerate(eeg_files):
    base_name = eeg_file.split('.')[0]
    nirs_match = [j for j, nirs_file in enumerate(nirs_files) if nirs_file.startswith(base_name)]
    if not nirs_match:
        continue
    j = nirs_match[0]
    if eeg_labels[i] != nirs_labels[j]:
        continue
    matched_data.append({
        'eeg': eeg_data[i],
        'nirs': nirs_data[j],
        'label': eeg_labels[i],
        'eeg_file': eeg_file,
        'nirs_file': nirs_files[j]
    })

print(f"\nNombre de paires EEG-NIRS valides : {len(matched_data)}")

eeg_data_matched = [item['eeg'] for item in matched_data]
nirs_data_matched = [item['nirs'] for item in matched_data]
labels = np.array([item['label'] for item in matched_data])

# =========================
#   Extraction de features
# =========================
def calculate_spectral_features(signal_data, sampling_rate=250):
    features = {}
    signal_data = signal_data.dropna()
    if len(signal_data) < 10:
        return {'spectral_power': 0, 'spectral_entropy': 0,
                'power_delta': 0, 'power_theta': 0, 'power_alpha': 0,
                'power_beta': 0, 'power_gamma': 0}
    try:
        freqs, psd = signal.welch(signal_data, fs=sampling_rate, nperseg=min(256, len(signal_data)//4))
        features['spectral_power'] = np.trapz(psd, freqs)
        psd_norm = psd / np.sum(psd) if np.sum(psd) > 0 else psd
        features['spectral_entropy'] = entropy(psd_norm) if np.sum(psd) > 0 else 0
        bands = {'delta': (0.5, 4),'theta': (4, 8),'alpha': (8, 13),'beta': (13, 30),'gamma': (30, 45)}
        for band, (fmin, fmax) in bands.items():
            idx = np.logical_and(freqs >= fmin, freqs <= fmax)
            band_power = np.trapz(psd[idx], freqs[idx]) if np.any(idx) else 0
            features[f'power_{band}'] = band_power
    except Exception:
        features = {'spectral_power': 0, 'spectral_entropy': 0,
                    'power_delta': 0, 'power_theta': 0, 'power_alpha': 0,
                    'power_beta': 0, 'power_gamma': 0}
    return features

def calculate_rms(signal_data):
    signal_data = signal_data.dropna()
    return float(np.sqrt(np.mean(signal_data**2))) if len(signal_data) > 0 else 0.0

def preprocess_eeg(eeg_data):
    eeg_processed = []
    for eeg in eeg_data:
        eeg_channels = [col for col in eeg.columns if col.startswith('EEG')]
        eeg_values = eeg[eeg_channels]

        def convert_value(x):
            try:
                x_str = str(x).strip()
                if x_str in ('-', ''):
                    return np.nan
                return float(x_str.replace(',', '.'))
            except:
                return np.nan

        eeg_values = eeg_values.map(convert_value)
        eeg_values = eeg_values.fillna(eeg_values.mean())

        features = []
        for channel in eeg_values.columns:
            ch = eeg_values[channel]
            # basiques
            features.extend([
                np.mean(ch), np.std(ch), np.min(ch), np.max(ch), np.median(ch), calculate_rms(ch)
            ])
            # spectrales
            sp = calculate_spectral_features(ch)
            features.extend([
                sp['spectral_power'], sp['spectral_entropy'],
                sp['power_delta'], sp['power_theta'], sp['power_alpha'], sp['power_beta'], sp['power_gamma']
            ])
        eeg_processed.append(features)
    return np.array(eeg_processed, dtype=float)

def extract_detector_features(df, detector_num):
    features = {}
    suffix = f'.{detector_num-1}' if detector_num > 1 else ''
    for channel in range(1, 17):
        red_col = f'RED{channel}{suffix}'
        ir_col = f'IR{channel}{suffix}'
        if red_col not in df.columns or ir_col not in df.columns:
            continue
        red_signal = pd.to_numeric(df[red_col], errors='coerce').fillna(method='bfill').fillna(method='ffill').values
        ir_signal  = pd.to_numeric(df[ir_col ], errors='coerce').fillna(method='bfill').fillna(method='ffill').values
        hbo = red_signal - ir_signal
        hbr = ir_signal - red_signal
        prefix = f'D{detector_num}C{channel}'
        features.update({
            f'{prefix}RED_mean': float(np.mean(red_signal)),
            f'{prefix}RED_std' : float(np.std(red_signal)),
            f'{prefix}IR_mean' : float(np.mean(ir_signal)),
            f'{prefix}HBO_mean': float(np.mean(hbo)),
            f'{prefix}HBR_mean': float(np.mean(hbr)),
        })
        diff_red = np.diff(red_signal)
        features[f'{prefix}RED_diff_std'] = float(np.std(diff_red)) if diff_red.size else 0.0
        if len(red_signal) > 10:
            f, psd = welch(red_signal, fs=10)
            features[f'{prefix}PSD_peak'] = float(np.max(psd)) if psd.size else 0.0
            features[f'{prefix}PSD_peak_freq'] = float(f[np.argmax(psd)]) if psd.size else 0.0
    return features

def preprocess_nirs(nirs_data):
    nirs_processed = []
    for nirs in nirs_data:
        features = []
        for detector in range(1, 33):
            detector_feats = extract_detector_features(nirs, detector)
            features.extend(list(detector_feats.values()))
        nirs_processed.append(features)
    return np.array(nirs_processed, dtype=float)

# Extraire
eeg_features  = preprocess_eeg(eeg_data_matched)
nirs_features = preprocess_nirs(nirs_data_matched)

def fuse_features(eeg_feats, nirs_feats):
    fused = []
    for eeg, nirs in zip(eeg_feats, nirs_feats):
        fused.append(np.concatenate([eeg, nirs]))
    return np.array(fused, dtype=float)

fused_features = fuse_features(eeg_features, nirs_features)

# =========================
#  Modèles classiques (dict)
# =========================
def classical_models_dict():
    return {
        'RandomForest': RandomForestClassifier(n_estimators=300, random_state=42),
        'LogisticRegression': Pipeline([
            ('imp', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler()),
            ('clf', LogisticRegression(max_iter=2000, random_state=42))
        ]),
        'MLP': Pipeline([
            ('imp', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler()),
            ('clf', MLPClassifier(hidden_layer_sizes=(256,128), activation='relu',
                                  max_iter=600, early_stopping=True, random_state=42))
        ]),
        'KNN': Pipeline([
            ('imp', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler()),
            ('clf', KNeighborsClassifier(n_neighbors=5))
        ]),
        'SVM': Pipeline([
            ('imp', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler()),
            ('clf', SVC(kernel='rbf', probability=True, random_state=42))
        ]),
        'LightGBM': LGBMClassifier(random_state=42, verbose=-1)
    }

# =========================
#  Modèles Deep: GRU & BiLSTM
# =========================
def build_gru(input_shape, lr=1e-3, dropout=0.2):
    model = Sequential([
        GRU(64, return_sequences=False, input_shape=input_shape),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dropout(dropout),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

def build_bilstm(input_shape, lr=1e-3, dropout=0.2):
    model = Sequential([
        Bidirectional(LSTM(64, return_sequences=False), input_shape=input_shape),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dropout(dropout),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# Helper: entraînement d'un modèle Keras sur un fold
def train_keras_cv(X_tr, y_tr, X_va, y_va, builder, epochs=60, batch_size=16):
    # imputation + scaling ➜ reshape en (n_samples, timesteps=n_features, 1)
    imp = SimpleImputer(strategy='mean')
    sc  = StandardScaler()

    X_tr_imp = imp.fit_transform(X_tr)
    X_va_imp = imp.transform(X_va)

    X_tr_scaled = sc.fit_transform(X_tr_imp)
    X_va_scaled = sc.transform(X_va_imp)

    X_tr_seq = X_tr_scaled.reshape((X_tr_scaled.shape[0], X_tr_scaled.shape[1], 1))
    X_va_seq = X_va_scaled.reshape((X_va_scaled.shape[0], X_va_scaled.shape[1], 1))

    model = builder(input_shape=(X_tr_seq.shape[1], 1))
    es = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=0)
    model.fit(X_tr_seq, y_tr, validation_data=(X_va_seq, y_va),
              epochs=epochs, batch_size=batch_size, callbacks=[es], verbose=0)
    y_pred_proba = model.predict(X_va_seq, verbose=0).ravel()
    y_pred = (y_pred_proba >= 0.5).astype(int)
    return y_pred, y_pred_proba
    
# =========================
#   Validation Croisée
# =========================
def cross_validate_all_models(X, y, modality_name="Modality", n_splits=5, random_state=42, plot_cm=True, only_best=False):
    print(f"\n===== Validation croisée ({n_splits}-fold) — {modality_name} =====")

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    models = classical_models_dict()

    # Stats accumulées
    scores = {name: {'acc':[], 'prec':[], 'rec':[], 'f1':[], 'roc_auc':[]} for name in list(models.keys()) + ['GRU','BiLSTM']}
    cms    = {name: np.zeros((2,2), dtype=int) for name in scores.keys()}

    fold_idx = 1
    for train_idx, val_idx in skf.split(X, y):
        print(f"- Fold {fold_idx}/{n_splits}")
        X_tr, X_va = X[train_idx], X[val_idx]
        y_tr, y_va = y[train_idx], y[val_idx]

        # --- Classiques ---
        for name, mdl in models.items():
            try:
                # LightGBM: réduire le logging
                if name == 'LightGBM':
                    reset_parameter()
                    log_evaluation(period=0)
                mdl.fit(X_tr, y_tr)
                y_pred = mdl.predict(X_va)
                y_pred_proba = mdl.predict_proba(X_va)[:, 1] if hasattr(mdl, 'predict_proba') else mdl.decision_function(X_va)
                cm = confusion_matrix(y_va, y_pred, labels=[0,1])
                cms[name] += cm
                scores[name]['acc'].append(accuracy_score(y_va, y_pred))
                scores[name]['prec'].append(precision_score(y_va, y_pred, zero_division=0))
                scores[name]['rec'].append(recall_score(y_va, y_pred, zero_division=0))
                scores[name]['f1' ].append(f1_score(y_va, y_pred, zero_division=0))
                scores[name]['roc_auc'].append(roc_auc_score(y_va, y_pred_proba))
            except Exception as e:
                print(f"  [!] {name} error: {e}")

        # --- GRU ---
        try:
            y_pred, y_pred_proba = train_keras_cv(X_tr, y_tr, X_va, y_va, build_gru)
            cm = confusion_matrix(y_va, y_pred, labels=[0,1])
            cms['GRU'] += cm
            scores['GRU']['acc'].append(accuracy_score(y_va, y_pred))
            scores['GRU']['prec'].append(precision_score(y_va, y_pred, zero_division=0))
            scores['GRU']['rec'].append(recall_score(y_va, y_pred, zero_division=0))
            scores['GRU']['f1' ].append(f1_score(y_va, y_pred, zero_division=0))
            scores['GRU']['roc_auc'].append(roc_auc_score(y_va, y_pred_proba))
        except Exception as e:
            print(f"  [!] GRU error: {e}")

        # --- BiLSTM ---
        try:
            y_pred, y_pred_proba = train_keras_cv(X_tr, y_tr, X_va, y_va, build_bilstm)
            cm = confusion_matrix(y_va, y_pred, labels=[0,1])
            cms['BiLSTM'] += cm
            scores['BiLSTM']['acc'].append(accuracy_score(y_va, y_pred))
            scores['BiLSTM']['prec'].append(precision_score(y_va, y_pred, zero_division=0))
            scores['BiLSTM']['rec'].append(recall_score(y_va, y_pred, zero_division=0))
            scores['BiLSTM']['f1' ].append(f1_score(y_va, y_pred, zero_division=0))
            scores['BiLSTM']['roc_auc'].append(roc_auc_score(y_va, y_pred_proba))
        except Exception as e:
            print(f"  [!] BiLSTM error: {e}")

        fold_idx += 1

    # Résumé
    def summarize(metric_list):
        return np.mean(metric_list), np.std(metric_list)

    summary_rows = []
    for name, m in scores.items():
        if len(m['acc']) == 0:
            continue
        acc_mu, acc_sd = summarize(m['acc'])
        pr_mu,  pr_sd  = summarize(m['prec'])
        rc_mu,  rc_sd  = summarize(m['rec'])
        f1_mu,  f1_sd  = summarize(m['f1'])
        roc_mu, roc_sd = summarize(m['roc_auc'])
        summary_rows.append({
            'Model': name,
            'Accuracy (μ±σ)': f"{acc_mu:.3f} ± {acc_sd:.3f}",
            'Precision (μ±σ)': f"{pr_mu:.3f} ± {pr_sd:.3f}",
            'Recall (μ±σ)':    f"{rc_mu:.3f} ± {rc_sd:.3f}",
            'F1 (μ±σ)':        f"{f1_mu:.3f} ± {f1_sd:.3f}",
            'ROC AUC (μ±σ)':   f"{roc_mu:.3f} ± {roc_sd:.3f}"
        })

    summary_df = pd.DataFrame(summary_rows).sort_values(by='Model').reset_index(drop=True)
    
    # Meilleur modèle sur la moyenne Accuracy
    best_name = None
    best_acc = -1
    for name, m in scores.items():
        if len(m['acc']) == 0: continue
        acc_mu = np.mean(m['acc'])
        if acc_mu > best_acc:
            best_acc = acc_mu
            best_name = name
    
    # Afficher seulement le meilleur modèle si only_best=True
    if only_best and best_name:
        best_row = summary_df[summary_df['Model'] == best_name]
        print(f"\nMeilleur modèle pour {modality_name}:")
        print(best_row.to_string(index=False))
        print(f"\nDétails du meilleur modèle ({best_name}):")
        print(f"  Accuracy: {np.mean(scores[best_name]['acc']):.3f} ± {np.std(scores[best_name]['acc']):.3f}")
        print(f"  Precision: {np.mean(scores[best_name]['prec']):.3f} ± {np.std(scores[best_name]['prec']):.3f}")
        print(f"  Recall: {np.mean(scores[best_name]['rec']):.3f} ± {np.std(scores[best_name]['rec']):.3f}")
        print(f"  F1-score: {np.mean(scores[best_name]['f1']):.3f} ± {np.std(scores[best_name]['f1']):.3f}")
        print(f"  ROC AUC: {np.mean(scores[best_name]['roc_auc']):.3f} ± {np.std(scores[best_name]['roc_auc']):.3f}")
    else:
        print("\n--- Résultats CV (moyenne ± écart-type) ---")
        print(summary_df.to_string(index=False))

    # Matrices de confusion cumulées
    if plot_cm and not only_best:
        n = len(cms)
        ncols = 3
        nrows = int(np.ceil(n / ncols))
        plt.figure(figsize=(5*ncols, 4*nrows))
        idx = 1
        for name, cm in cms.items():
            if cm.sum() == 0:  # modèle qui a échoué
                continue
            plt.subplot(nrows, ncols, idx)
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                        xticklabels=['SO','AO'], yticklabels=['SO','AO'])
            plt.title(f"{modality_name} — {name}")
            plt.xlabel("Prédit"); plt.ylabel("Vrai")
            idx += 1
        plt.tight_layout()
        plt.show()

    return summary_df, cms, scores, best_name

# =========================
#   Lancement CV par modalité
# =========================
# EEG seul (afficher seulement le meilleur)
cv_eeg_df, cv_eeg_cm, _, best_eeg = cross_validate_all_models(
    eeg_features, labels, modality_name="EEG seul", n_splits=5, only_best=True
)

# NIRS seul (afficher seulement le meilleur)
cv_nirs_df, cv_nirs_cm, _, best_nirs = cross_validate_all_models(
    nirs_features, labels, modality_name="NIRS seul", n_splits=5, only_best=True
)

# Fusion (afficher tous les modèles)
cv_fused_df, cv_fused_cm, _, _ = cross_validate_all_models(
    fused_features, labels, modality_name="Fusion EEG+NIRS", n_splits=5, only_best=False
)
```

```python
# =========================
#  Histogramme 
# =========================
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PALETTE = {
    "Accuracy":  "#1f77b4",
    "Precision": "#2ca02c",
    "Recall":    "#8c564b",
    "F1-Score":  "#7f7f7f",
    "ROC-AUC":   "#17becf",
}

def _parse_mu_sigma(series: pd.Series):
    s = series.astype(str).str.strip()
    mu = s.str.split('±').str[0].str.replace(r'[^0-9eE+\-\.]', '', regex=True)
    mu = pd.to_numeric(mu, errors='coerce').fillna(0.0).values
    return mu

def plot_models_hist_style(summary_df: pd.DataFrame,
                           title="EEG + NIRS + Fusion CV",
                           model_order=None):
    # Détection colonnes
    cols = {c.lower(): c for c in summary_df.columns}
    def find(key, regex=False):
        for k,v in cols.items():
            if (not regex and key in k) or (regex and re.search(key, k)):
                return v
        return None
    c_model = find('model'); c_acc = find('accuracy'); c_prec = find('precision')
    c_rec = find('recall');  c_f1  = find(r'\bf1\b', regex=True); c_roc = find('roc')
    if any(x is None for x in [c_model,c_acc,c_prec,c_rec,c_f1,c_roc]):
        raise ValueError("Colonnes attendues manquantes (Model, Accuracy, Precision, Recall, F1, ROC).")

    df = summary_df.copy()
    if model_order:
        df['_ord'] = df[c_model].apply(lambda x: model_order.index(x) if x in model_order else 10_000)
        df = df.sort_values('_ord').drop(columns=['_ord'])

    models  = df[c_model].tolist()
    metrics = ["Accuracy","Precision","Recall","F1-Score","ROC-AUC"]

    mu_acc = _parse_mu_sigma(df[c_acc])
    mu_prc = _parse_mu_sigma(df[c_prec])
    mu_rec = _parse_mu_sigma(df[c_rec])
    mu_f1  = _parse_mu_sigma(df[c_f1])
    mu_roc = _parse_mu_sigma(df[c_roc])
    mus = [mu_acc, mu_prc, mu_rec, mu_f1, mu_roc]

    # Barres groupées (sans yerr)
    x = np.arange(len(models))
    width = 0.14
    offsets = np.linspace(-2, 2, 5) * width  # 5 métriques

    plt.figure(figsize=(18,7))
    ax = plt.gca()
    ax.set_axisbelow(True)  # la grille passe derrière

    for i, metric in enumerate(metrics):
        y = mus[i]
        plt.bar(x + offsets[i], y, width,
                color=PALETTE[metric],
                edgecolor=PALETTE[metric], linewidth=0,   # aucun contour sombre
                label=metric, zorder=3)                   # barres au-dessus de la grille

    # Style
    for spine in ax.spines.values():
        spine.set_color('#B5B5B5')
    plt.tick_params(axis='both', colors='#666666')
    ax.yaxis.grid(True, linestyle='--', linewidth=0.6, color='#DDDDDD')
    ax.xaxis.grid(False)

    plt.xticks(x, models, rotation=30, ha='right', color='#444444')
    plt.ylabel("Score (0 to 1)", color='#444444')
    plt.xlabel("Models", color='#444444')
    plt.title(title, color='#444444')
    plt.ylim(0, 1.02)
    leg = plt.legend(title="Metrics", loc='center left', bbox_to_anchor=(1.015, 0.5), frameon=True)
    lf = leg.get_frame(); lf.set_edgecolor('#B5B5B5'); lf.set_facecolor('white'); lf.set_alpha(0.95)
    plt.margins(x=0.02)
    plt.tight_layout()
    plt.show()

# Ordre identique à ta capture si souhaité :
order = ['SVM','XGBoost','BiLSTM','GRU','KNN','Random Forest','MLP','Logistic Regression','LightGBM']
plot_models_hist_style(cv_fused_df,
                       title="EEG + NIRS + Fusion CV (Classiques + GRU + BiLSTM)",
                       model_order=order)
```
